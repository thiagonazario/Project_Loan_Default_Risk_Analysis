```python
# Vamos escrever uma função que calcule a categoria de idade
def age_category(age):
    if age < 20:
        return 'Adolescent'
    elif age < 35:
        return 'Young adult'
    elif age < 50:
        return 'Adult'
    elif age < 65:
        return 'Middle-aged Adult'
    else:
        return 'Senior citizen / Elderly'
```

```python
# Teste se a função funciona
for age in df['dob_years']:
    category = age_category(age)
    print(f'A categoria para a idade de {age} anos é: {category}')
```

```python
# Criar coluna nova com base na função
df['age_category'] = df['dob_years'].apply(age_category)
```

```python
# categorizando por número de filhos
def children_status(df):
    df.loc[df['children'] == 0, 'children_status'] = 'no child'
    df.loc[df['children'] >= 3, 'children_status'] = '3 or more children'
    df.loc[(df['children'] > 0) & (df['children'] < 3), 'children_status'] = '1-2 children'
    
    return df
```

```python
# criando categorias de renda
def income_class(df):
    df.loc[df['total_income'] >= 40000, 'income_class'] = 'high class'
    df.loc[df['total_income'] <= 20000, 'income_class'] = 'low class'
    df.loc[(df['total_income'] > 20000) & (df['total_income'] < 40000), 'income_class'] = 'middle class'
    
    return df
```
